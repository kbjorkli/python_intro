# Matplotlib

*Matplotlib* er en pakke som lar oss presentere data i Python grafisk, et såkalt "plott". Særlig bruker vi Matplotlib til å lage matematiske grafer av arrayer. Vi importerer underpakken *pyplot* (som inneholder nærmest all funkjsonalitet vi trenger), vanligvis med navnet 'plt', med syntaksen *matplotlib.pyplot as plt*.

### Eksempel

```python
import matplotlib.pyplot as plt
```

## Grunnleggende plotting

Vi kan bruke Matplotlib-funksjonene *plot* og *show* for å henholdvis tegne en graf og vise et plott. plot-funksjonen tar to (like store!) arrayer/lister som argumenter og behandler dem som x- og y-verdier til en rekke punkter. Den tegner en graf ved å trekke en rett linje mellom alle disse punktene, med syntaksen *plot(x_verdier, y_verdier)*. show-funksjonen er den som faktisk viser plottet - uten den vil du ikke få se resultatet!

### Eksempel

```python
import matplotlib.pyplot as plt
import numpy as np

x_verdier = np.linspace(0,2*np.pi,10)
y_verdier = np.sin(x_verdier)

plt.plot(x_verdier, y_verdier)  # punktene i grafen blir da (x_verdier[0], y_verdier[0]), (x_verdier[1], y_verdier[1]) osv.
plt.show()
```

Vi kan plotte flere grafer i samme plott ved å kalle plot-funksjonen flere ganger før vi kaller show. Vi kan også forbedre "oppløsningen" til grafen ved å bruke flere punkter.

### Eksempel

```python
import matplotlib.pyplot as plt
import numpy as np

x_verdier = np.linspace(0,2*np.pi,50)  # Nå med 50 elementer i stedet for 10!
y_verdier = np.sin(x_verdier)
y2_verdier = np.cos(x_verdier)

plt.plot(x_verdier, y_verdier)
plt.plot(x_verdier, y2_verdier)
plt.show()
```

Vi kan endre på utseende til grafene ved å gi ekstraargumenter i plot-funksjonen.

### Eksempel

```python
import matplotlib.pyplot as plt
import numpy as np

x_verdier = np.linspace(0,2*np.pi,50)
y_verdier = np.sin(x_verdier)
y2_verdier = np.cos(x_verdier)

plt.plot(x_verdier, y_verdier, '--', color='green')
plt.plot(x_verdier, y2_verdier, '.', color='purple')
plt.show()
# mange flere varianter finner du i Matplotlib-dokumentasjonen på nett
```

## Plott-tilbehør

Vi kan utbrodere plottet vårt med bruk av yttligere funksjoner fra Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np

x_verdier = np.linspace(0,2*np.pi,50)
y_verdier = np.sin(x_verdier)

plt.plot(x_verdier, y_verdier, label = "sin(x)")

plt.title("Sinusfunksjonen")  # gir en tittel til plottet
plt.grid()                    # lager et rutenett i plottet
plt.xlabel("x")               # argumentet gir navn til x-aksen
plt.ylabel("f(x)")
plt.legend()                  # viser label = "..."-argumentet fra plotfunksjonen

plt.show()
```
